# Example for hillshading / Schummerung


```python
import numpy as np
from matplotlib.colors import LightSource
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1, 1, 101)
xx = np.outer(x, x)
z = np.sinc(xx)
```

```python
plt.imshow(z)
```

```python
azdeg = 35. # azimuth of light source
altdeg = 20. # altitude of light source
ve = 1. # vertical exaggeration
ls = LightSource(azdeg=azdeg, altdeg=altdeg)
rgb = ls.shade(z, cmap=plt.cm.gray, vert_exag=ve, blend_mode='soft')
plt.imshow(rgb, aspect='auto')
```

```python
plt.imshow(z)
plt.imshow(rgb, alpha=0.4)
```
